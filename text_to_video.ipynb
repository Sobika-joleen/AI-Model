```python
# 🧱 Install required libraries
!pip install diffusers transformers accelerate scipy safetensors
!pip install moviepy gtts

# 📦 Import libraries
from diffusers import StableDiffusionPipeline
import torch, os
from PIL import Image
from moviepy.editor import ImageSequenceClip, AudioFileClip
from gtts import gTTS
from IPython.display import Video

# 🎤 Get input from the user
prompt = input("Enter your prompt for image generation: ")
narration_text = input("Enter narration text for the video: ")

# 🚀 Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2",
    torch_dtype=torch.float16,
    safety_checker=None,
    use_safetensors=True
).to("cuda")

# 🖼️ Generate images
NUM_FRAMES = 5
FRAME_DIR = "frames"
os.makedirs(FRAME_DIR, exist_ok=True)

for i in range(NUM_FRAMES):
    image = pipe(prompt).images[0]
    image.save(f"{FRAME_DIR}/frame{i}.png")

# 🎬 Create video from images
image_files = [f"{FRAME_DIR}/frame{i}.png" for i in range(NUM_FRAMES)]
clip = ImageSequenceClip(image_files, fps=1)
clip.write_videofile("video.mp4")

# 🔊 Convert narration text to speech
gtts = gTTS(narration_text)
gtts.save("narration.mp3")

# 🎞️ Combine audio with video
final = clip.set_audio(AudioFileClip("narration.mp3"))
final.write_videofile("final_video.mp4")

# 📽️ Display the video
Video("final_video.mp4", embed=True)
```

```python
!pip install pyttsx3
!pip install gTTS
!pip install ipywidgets
from gtts import gTTS
from IPython.display import Audio, display
import ipywidgets as widgets
from google.colab import files

# Text input widget
text_input = widgets.Textarea(
    value='Hello! This is a dynamic Text to Speech converter.',
    placeholder='Type your text here...',
    description='Text:',
    layout=widgets.Layout(width='100%', height='100px')
)

# Language dropdown widget
language_dropdown = widgets.Dropdown(
    options=[('English', 'en'), ('Hindi', 'hi'), ('French', 'fr'), ('Spanish', 'es'), ('Tamil', 'ta')],
    value='en',
    description='Language:'
)

# Speed toggle widget
speed_toggle = widgets.ToggleButtons(
    options=[('Normal', False), ('Slow', True)],
    description='Speed:',
    button_style=''
)

# Button to generate TTS
generate_button = widgets.Button(
    description='Generate Speech',
    button_style='success',
    tooltip='Click to generate speech',
    icon='play'
)

# Output widget to display audio and messages
output = widgets.Output()

# Function to handle button click
def generate_speech(b):
    with output:
        output.clear_output()  # Clear previous output
        text = text_input.value
        language = language_dropdown.value
        slow = speed_toggle.value

        if not text.strip():
            print("⚠️ Please enter some text.")
            return

        try:
            # Generate speech
            tts = gTTS(text=text, lang=language, slow=slow)
            tts.save("output.mp3")
            print("✅ Speech generated successfully! Play or download below ⬇️")

            # Play audio
            display(Audio("output.mp3"))

            # Download button
            files.download('output.mp3')
        except Exception as e:
            print(f"❌ Error: {e}")

# Attach event to button
generate_button.on_click(generate_speech)

# Display all widgets
display(text_input, language_dropdown, speed_toggle, generate_button, output)
```
